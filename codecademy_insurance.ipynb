```python
#create a column using a lambda function that separates the 
#bmi data in three categories, then graph age and cost with 
#three categories of bmi

df['bmi_category'] = df.bmi.apply(
    lambda x: 'Overweight' if x >= 25 else ('Underweight' if x < 18.5 else 'Normal'))
```

```python
print(df.bmi_category.unique())
```

```python
sns.scatterplot(x = df.age, y = df.charges, hue = df.bmi_category, palette='bright')
```
